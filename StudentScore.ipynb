```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_score.csv")
#provide top 5 data og each column with Heading
print(df.head())
```

```python
df.describe()
```

```python
#to check what are the datatypes
df.info() 
```

```python
#to check how many values are null
df.isnull().sum()
```

```python
#drop unnamed column ie delete it
df = df.drop("Unnamed: 0", axis = 1)
```

```python
print(df.head())
```

```python
#Gender Distribution
plt.figure(figsize =(5,5))
ax = sns.countplot(data = df ,x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analyze : the number of female in the data is more than number of male
```

```python
#Parents Education score
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize =(5,5))
sns.heatmap(gb,annot= True)
plt.title("Relationship between Parent's score and Students Score")
plt.show()
```

```python
#from the above chart we have analyze the education of parents has good impact on the students score
```

```python
#Parents MaritalStatus
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize =(5,5))
sns.heatmap(gb1,annot= True)
plt.title("Relationship between Parent's Marital Status and Students Score")
plt.show()
```

```python
#from the above chart we have analyze the marital status dont impact the score of students
```

```python
#check how many groups are their
print(df["EthnicGroup"].unique())
```

```python
#Distribution of Ethinic Group
groupA= df.loc[(df['EthnicGroup'] == "group A")].count()
groupB= df.loc[(df['EthnicGroup'] == "group B")].count()
groupC= df.loc[(df['EthnicGroup'] == "group C")].count()
groupD= df.loc[(df['EthnicGroup'] == "group D")].count()
groupE= df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist ,labels =l, autopct = "%1.2f%%")
plt.title("Disctribution of Ethnic Groups")
plt.show()
```
